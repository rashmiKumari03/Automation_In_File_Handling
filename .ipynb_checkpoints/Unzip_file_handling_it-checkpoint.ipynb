### Step 1: This is the way to unzip the folder...and save entire unzip to destination or new path.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import zipfile
import os

def unzip_folder(zip_path, extract_to):   
    
    # Creating function ...to unzip the folder and save the unzip file to new destination.
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
```

### Step 2: Reading all the txt files from  each subfolders.

```python
def get_txt_files(directory):
    txt_files = []
    for root, dirs, files in os.walk(directory):
        print(root)
        print(dirs)
        print(files)
        for file in files:
            if file.endswith(".txt"):
                txt_files.append(os.path.join(root, file))
    return txt_files
```

```python
txt_files_list = get_txt_files(extract_folder_path)

print("Extracted .txt files:")
for txt_file in txt_files_list:
    print(txt_file)
```

### Step : This is the step to convert the txt file into csv file and save them all in new folder called Data...(create one if not there)


### Since we have tab in the txt file (values separated by tabs) and so they are kind of arranged....

```python
os.getcwd()
```

```python
import csv
import os

output_folder = 'Data'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

for text_file in txt_files_list:
    
    # Read text file and extract data
    with open(text_file, 'r') as file:
        data = [line.strip().split('\t') for line in file.readlines()]

    # Create CSV file name
    csv_file = os.path.join(output_folder, os.path.splitext(os.path.basename(text_file))[0] + '.csv')
    print("Text File is     : ",os.path.basename(text_file))
    
    print("Created CSV File is :",csv_file)
    print("----------------------------------")

    # Write data to CSV file
    with open(csv_file, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerows(data)

print("Conversion complete!!!! CSV files saved in the 'Data' folder.")
```

```python
import pandas as pd
import glob

# Path to the directory containing your CSV files
csv_files_path = 'Data/'

# Get a list of all CSV files in the directory
csv_files = glob.glob(csv_files_path + '*.csv')
```

```python
concatenated_2018_I1 = pd.DataFrame()
concatenated_2018_I2 = pd.DataFrame()
concatenated_2018_MFM1 = pd.DataFrame()
concatenated_2018_WMS = pd.DataFrame()

concatenated_2019_I1 = pd.DataFrame()
concatenated_2019_I2 = pd.DataFrame()
concatenated_2019_MFM1 = pd.DataFrame()
concatenated_2019_WMS = pd.DataFrame()

for i in csv_files:
    data = pd.read_csv(i)
    print(f"\nFile: {i}")
    print(f"Original Shape: {data.shape}")
    print(f"Columns: {data.columns}")

    if '2018' in i:
        if "I1" in i:
            concatenated_2018_I1 = pd.concat([concatenated_2018_I1, data], ignore_index=True)
        elif "I2" in i:
            concatenated_2018_I2 = pd.concat([concatenated_2018_I2, data], ignore_index=True)
        elif "MFM1" in i:
            concatenated_2018_MFM1 = pd.concat([concatenated_2018_MFM1, data], ignore_index=True)
        elif "WMS" in i:
            concatenated_2018_WMS = pd.concat([concatenated_2018_WMS, data], ignore_index=True)

    elif '2019' in i:
        if "I1" in i:
            concatenated_2019_I1 = pd.concat([concatenated_2019_I1, data], ignore_index=True)
        elif "I2" in i:
            concatenated_2019_I2 = pd.concat([concatenated_2019_I2, data], ignore_index=True)
        elif "MFM1" in i:
            concatenated_2019_MFM1 = pd.concat([concatenated_2019_MFM1, data], ignore_index=True)
        elif "WMS" in i:
            concatenated_2019_WMS = pd.concat([concatenated_2019_WMS, data], ignore_index=True)

    print(f"Concatenated Shape (2018):")
    print("Number of rows and columns in I1:", concatenated_2018_I1.shape)
    print("Number of rows and columns in I2:", concatenated_2018_I2.shape)
    print("Number of rows and columns in MFM1:", concatenated_2018_MFM1.shape)
    print("Number of rows and columns in WMS:", concatenated_2018_WMS.shape)
    print("---------------------------------------------------------------------------------")
    print(f"Concatenated Shape (2019):")
    print("Number of rows and columns in I1:", concatenated_2019_I1.shape)
    print("Number of rows and columns in I2:", concatenated_2019_I2.shape)
    print("Number of rows and columns in MFM1:", concatenated_2019_MFM1.shape)
    print("Number of rows and columns in WMS:", concatenated_2019_WMS.shape)
```

```python
output_directory = "Concated_Datasets"
os.makedirs(output_directory, exist_ok=True)
```

```python
# Save the concatenated data to a new CSV file

concatenated_2018_I1.to_csv(os.path.join(output_directory,'Concatenated_2018_I1.csv'), index=False)
concatenated_2018_I2.to_csv(os.path.join(output_directory,'Concatenated_2018_I2.csv'), index=False)
concatenated_2018_MFM1.to_csv(os.path.join(output_directory,'Concatenated_2018_MFM1.csv'), index=False)
concatenated_2018_WMS.to_csv(os.path.join(output_directory,'Concatenated_2018_WMS.csv'), index=False)


concatenated_2019_I1.to_csv(os.path.join(output_directory,'Concatenated_2019_I1.csv'), index=False)
concatenated_2019_I2.to_csv(os.path.join(output_directory,'Concatenated_2019_I2.csv'), index=False)
concatenated_2019_MFM1.to_csv(os.path.join(output_directory,'Concatenated_2019_MFM1.csv'), index=False)
concatenated_2019_WMS.to_csv(os.path.join(output_directory,'Concatenated_2019_WMS.csv'), index=False)
```

## How can we automate featching the names like I1,I2,MFM1 and WMS from both 2018 and 2019 dataset..

```python
for i in csv_files:
    
    parts = i.split('-')
    print(parts)
    
```

```python
# Create an empty dictionary to store DataFrames
concatenated_data = {}

# Iterate through the file names
for i in csv_files:
    
    # Extract information from the file name
    parts = i.split('-')

    category = parts[2]
    # print(category)
    
    year= parts[3]
    # print(year)


    # Create DataFrame name based on the information extracted
    df_name = f"Automated_concatenated_{year}_{category}"
    print(df_name)

    # If the DataFrame does not exist in the dictionary, create it
    if df_name not in concatenated_data:
        concatenated_data[df_name] = pd.DataFrame()   #To avoid repreation

    # Concatenate the data to the corresponding DataFrame
    data = pd.read_csv(i)
    concatenated_data[df_name] = pd.concat([concatenated_data[df_name],data], ignore_index=True)

# Print the shapes of the concatenated DataFrames
for df_name, df in concatenated_data.items():
    print("Number of rows and columns:",df.shape)
    print("Columns Name:",df.columns)
    print("---------------------------------------------------------------------------------")
```

```python
output_directory = "Automated_Concated_Datasets"
os.makedirs(output_directory, exist_ok=True)
```

```python
# Save the concatenated data to a new CSV file

for df_name,df in concatenated_data.items():    # As concatenated_data is dict so get keys and values using items()
    print(df_name)

    # Was a dict concatenated_data having dataframes..
    
    # Automated_concatenated_2018_I1 is the df_name...
    # so df_name.split('_')[2] gives year 
    # and df_name.split('_')[3] gives category
    
    year, category = df_name.split('_')[2],df_name.split('_')[3]
    
    #Create the csv file....for 2018 and 2019
    output_file = f"Automated_Concatenated_{year}_{category}.csv"
   
    output_path = os.path.join(output_directory, output_file)
    df.to_csv(output_path, index=False)
```
